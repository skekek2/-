```python
x = np.arange(10)
fig, ax = plt.subplots()

# 1. label 뒤에 쉼표(,)를 추가했습니다.
ax.plot(x, x, label='y=x', linestyle='-', marker='.', color='blue')

# 2. linestyle의 따옴표('-')를 닫아주고 맨 끝의 오타 'I'를 지웠습니다.
ax.plot(x, x ** 2, label='y=x^2', linestyle='-', marker='s', color='red')

# 그래프 확인 및 저장
plt.legend() # 범례(label)를 화면에 표시하기 위해 추가하면 좋습니다.
plt.show()   # 코랩에서 그래프를 즉시 확인하기 위해 추가
fig.savefig("plot.png")
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
plt.title('Chart Test') # 차트 제목

# 1. color 앞에 쉼표 추가 및 맨 끝에 닫는 괄호 ) 추가
ax.plot(x, x, label='y=x', linestyle='-', marker='.', color='blue')

# 2. 맨 끝에 닫는 괄호 ) 추가
ax.plot(x, x ** 2, label='y=x^2', linestyle='-.', marker=',', color='red')

ax.set_xlabel("x") # X축의 제목
ax.set_ylabel("y") # y축의 제목

# 3. 맨 끝에 닫는 괄호 ) 추가
ax.legend(loc='best', shadow=False, fancybox=True, borderpad=2)

# 그래프 확인 및 저장
plt.show()
fig.savefig("plot.png")
```

```python
# 1. 도화지(fig, axes)를 가장 먼저 만듭니다.
plt.rc('font', family='NanumBarunGothic')
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# 2. 데이터 세트 준비 (첫 줄의 "축구"' 오타도 고쳐두었습니다)
x = np.array(["축구", "야구", "농구", "배드민턴", "탁구"])
y = np.array([13, 10, 17, 8, 7])
z = np.random.randn(1000)

# 3. 먼저 만든 도화지(axes)에 그래프를 그립니다.
axes[0].bar(x, y)       # 왼쪽 상자에 Bar 그래프
axes[1].hist(z, bins=50) # 오른쪽 상자에 히스토그램

# 4. 그림이 다 그려진 도화지를 마지막에 저장하고 화면에 출력합니다.
fig.savefig("plot.png")
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]

# 1. A에 따옴표를 붙이고, 변수 이름을 labels로 통일했습니다.
labels = ['A', 'B', 'C', 'D']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']

plt.axis('equal') # 동그란 원 모양 유지

# 2. plt.ple를 plt.pie로 고치고 labels 변수를 올바르게 매칭했습니다.
plt.pie(size, labels=labels, autopct='%.1f%%', explode=(0.1, 0.1, 0.1, 0.1), colors=color)

plt.legend()
plt.show()
```
