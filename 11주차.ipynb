```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 한글 폰트 설정 (가장 안전한 직접 지정 방식)
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 2. 데이터 정렬 및 Top10 추출
# (⚠️ 주의: 세션을 다시 시작했으므로, 이 코드 윗줄이나 이전 셀에서 df = pd.read_csv(...) 등으로 데이터를 먼저 불러와야 합니다!)
top10_off = df.sort_values(by='평균 하차 인원 수', ascending=False).head(10)

# 3. 그래프 객체 생성
fig, ax = plt.subplots(figsize=(20, 10))

# 4. 막대그래프 그리기
# X축은 데이터프레임의 index(역 이름), Y축은 '평균 하차 인원 수'
bars = ax.bar(top10_off.index, top10_off['평균 하차 인원 수'])

# 5. 각 막대 위에 숫자(텍스트) 표시하기
for i, bar in enumerate(bars):
    yval = bar.get_height()

    # 천 단위 쉼표 추가 (예: 1234560 -> 1,234,560)
    label_text = f'{int(yval):,}'

    # 1등 역은 빨간색 + 굵게, 나머지는 검은색
    text_color = 'red' if i == 0 else 'black'
    font_weight = 'bold' if i == 0 else 'normal'

    # 막대 위 정중앙에 텍스트 배치
    ax.text(bar.get_x() + bar.get_width()/2, yval, label_text,
            va='bottom', ha='center', color=text_color, fontsize=12, fontweight=font_weight)

# 6. 제목 및 축 레이블에 한글 폰트 개별 적용 (버그 방지)
ax.set_title('2023년 10월 평균 하차 인원 수 Top10', fontproperties=font_prop, fontsize=18)
ax.set_xticklabels(top10_off.index, fontproperties=font_prop, fontsize=12, rotation=45)
ax.set_ylabel('인원 수', fontproperties=font_prop, fontsize=14)

# 7. 그래프 강제 출력
plt.show()
```

```python
# 지하철 역별 위치좌표에 대한 데이터를 물러읍니다.
subway_location = pd.read_csv('지하철 역 위치 좌표.csv')
subway_location
```

```python
def get_nums_and_location(line, metro_st):
    # 특정 호선의 데이터만 추출합니다.
    metro_line_n = metro_st[metro_st['호선명'] == line]

    # 💡 여기서부터 모든 코드의 앞에 들여쓰기(Tab 또는 공백 4칸)를 추가합니다!
    # 승차 인원 컬럼만 추출합니다.
    metro_get_on = pd.DataFrame()
    metro_get_on['지하철역'] = metro_line_n['지하철역']
    for i in range(int((len(metro_recent.columns)-3)/2)):
        metro_get_on[metro_line_n.columns[3+2*i]] = metro_line_n[metro_line_n.columns[3+2*i]]
    metro_get_on = metro_get_on.set_index('지하철역')

    # 하차 인원 컬럼만 추출합니다.
    metro_get_off = pd.DataFrame()
    metro_get_off['지하철역'] = metro_line_n['지하철역']
    for i in range(int((len(metro_recent.columns)-3)/2)):
        metro_get_off[metro_line_n.columns[4+2*i]] = metro_line_n[metro_line_n.columns[4+2*i]]
    metro_get_off = metro_get_off.set_index('지하철역')

    # 역 별 평균 승하차 인원을 구한 후 정수로 형 변환하여 데이터프레임으로 저장합니다.
    df = pd.DataFrame(index = metro_line_n['지하철역'])
    df['평균 승차 인원 수'] = metro_get_on.mean(axis=1).astype(int)
    df['평균 하차 인원 수'] = metro_get_off.mean(axis=1).astype(int)

    # 지하철역 명 동일하도록 설정합니다.
    temp = []
    df = df.reset_index()
    for name in df['지하철역']:
        temp.append(name.split('(')[0]+'')
    df['지하철역'] = temp

    # 지하철역 명을 기준으로 두 데이터프레임 병합합니다.
    df = df.merge(subway_location, left_on='지하철역', right_on='지하철역')

    return df
```
